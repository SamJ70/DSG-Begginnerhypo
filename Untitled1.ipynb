```python
import os
import json
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from torch.optim.lr_scheduler import StepLR
from torch.cuda.amp import GradScaler, autocast
from PIL import Image

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Custom Dataset Class
class SurgicalTripletDataset(Dataset):
    def __init__(self, videos_dir, labels_dir, transform=None):
        self.videos_dir = videos_dir
        self.labels_dir = labels_dir
        self.transform = transform
        self.video_ids = [vid for vid in os.listdir(videos_dir) if os.path.isdir(os.path.join(videos_dir, vid))]
        self.frames = self._load_frames()

    def _load_frames(self):
        frames = []
        for video_id in self.video_ids:
            video_path = os.path.join(self.videos_dir, video_id)
            frame_files = sorted(os.listdir(video_path))
            for frame_file in frame_files:
                frames.append((video_id, frame_file))
        return frames

    def __len__(self):
        return len(self.frames)

    def __getitem__(self, idx):
        video_id, frame_file = self.frames[idx]
        frame_path = os.path.join(self.videos_dir, video_id, frame_file)
        label_path = os.path.join(self.labels_dir, f"{video_id}.json")

        # Load frame
        frame = Image.open(frame_path).convert("RGB")
        if self.transform:
            frame = self.transform(frame)

        # Load labels
        with open(label_path, "r") as f:
            labels = json.load(f)

        # Extract frame ID from filename (e.g., "000000.png" -> "0")
        frame_id = str(int(frame_file.split(".")[0]))  # Convert to integer first, then to string

        # Get labels for the current frame from the "annotations" key
        frame_labels = labels.get("annotations", {}).get(frame_id, [])

        # Skip frames without labels
        if not frame_labels:
            return None, None

        # Parse the labels (assuming the format is [ID, SC, BX, BY, BW, BH, Instrument, verb, target, phase])
        instrument = frame_labels[0][6]  # Instrument class
        verb = frame_labels[0][7]       # Verb class
        target = frame_labels[0][8]     # Target class
        bbox = frame_labels[0][2:6]     # Bounding box coordinates (BX, BY, BW, BH)

        # Handle missing bounding box information
        if any(coord == -1 for coord in bbox):
            bbox = [0, 0, 0, 0]  # Default bounding box if missing

        # Convert bounding box to tensor
        bbox = torch.tensor(bbox, dtype=torch.float32)

        # Return frame and parsed labels
        return frame, {
            "instrument": instrument,
            "verb": verb,
            "target": target,
            "bbox": bbox,
        }

# Model Architecture
class SurgicalActionModel(nn.Module):
    def __init__(self, num_instruments, num_verbs, num_targets):
        super(SurgicalActionModel, self).__init__()
        # Load ResNet-50 backbone
        self.backbone = models.resnet50(pretrained=True)

        # Remove the final classification layer and global average pooling
        self.backbone = nn.Sequential(*list(self.backbone.children())[:-2])  # Keep the convolutional layers

        # Triplet recognition heads
        self.instrument_head = nn.Linear(2048, num_instruments)  # Ensure num_instruments matches the dataset
        self.verb_head = nn.Linear(2048, num_verbs)              # Ensure num_verbs matches the dataset
        self.target_head = nn.Linear(2048, num_targets)          # Ensure num_targets matches the dataset

        # Localization head
        self.localization_head = nn.Sequential(
            nn.Conv2d(2048, 512, kernel_size=3, padding=1),  # Reduce channels from 2048 to 512
            nn.ReLU(),
            nn.Conv2d(512, 4, kernel_size=1),  # Predict bounding box coordinates (x, y, w, h)
        )

    def forward(self, x):
        # Extract features from the backbone
        features = self.backbone(x)  # Shape: [batch_size, 2048, height, width]

        # Global average pooling for triplet recognition
        pooled_features = torch.mean(features, dim=[2, 3])  # Shape: [batch_size, 2048]

        # Triplet recognition
        instrument_logits = self.instrument_head(pooled_features)
        verb_logits = self.verb_head(pooled_features)
        target_logits = self.target_head(pooled_features)

        # Localization
        bbox = self.localization_head(features)  # Shape: [batch_size, 4, height, width]
        bbox = torch.mean(bbox, dim=[2, 3])      # Global average pooling to get [batch_size, 4]

        return instrument_logits, verb_logits, target_logits, bbox

# Focal Loss for imbalanced classes
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, ignore_index=-1):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.ignore_index = ignore_index

    def forward(self, inputs, targets):
        # Mask out invalid targets (e.g., -1 or out-of-bounds values)
        valid_mask = (targets != self.ignore_index) & (targets < inputs.size(1))
        inputs = inputs[valid_mask]
        targets = targets[valid_mask]

        if inputs.numel() == 0:  # If all targets are invalid, return 0 loss
            return torch.tensor(0.0, device=inputs.device)

        # Compute focal loss
        BCE_loss = nn.functional.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-BCE_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * BCE_loss
        return focal_loss.mean()

# Define paths
dataset_path = "/content/CholecT50"  # Updated dataset path
videos_dir = os.path.join(dataset_path, "videos")
labels_dir = os.path.join(dataset_path, "labels")

# Define transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Create dataset and dataloader
dataset = SurgicalTripletDataset(videos_dir, labels_dir, transform=transform)
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# Collect all unique values for instrument, verb, and target
instruments = set()
verbs = set()
targets = set()

for _, labels in dataset:
    instruments.add(labels["instrument"])
    verbs.add(labels["verb"])
    targets.add(labels["target"])

# Find the maximum values
max_instrument = int(max(instruments))
max_verb = int(max(verbs))
max_target = int(max(targets))

print(f"Max instrument: {max_instrument}")
print(f"Max verb: {max_verb}")
print(f"Max target: {max_target}")

# Initialize model, optimizer, and loss functions
model = SurgicalActionModel(
    num_instruments=max_instrument + 1,  # +1 to account for 0-based indexing
    num_verbs=max_verb + 1,
    num_targets=max_target + 1,
).to(device)

optimizer = optim.Adam(model.parameters(), lr=1e-4)
scheduler = StepLR(optimizer, step_size=5, gamma=0.1)
scaler = GradScaler()

# Focal Loss for each task
focal_loss_instrument = FocalLoss(ignore_index=-1)
focal_loss_verb = FocalLoss(ignore_index=-1)
focal_loss_target = FocalLoss(ignore_index=-1)

# Training loop
for epoch in range(7):
    for frames, labels in dataloader:
        # Move data to GPU
        frames = frames.to(device)
        labels["instrument"] = labels["instrument"].to(device).long()
        labels["verb"] = labels["verb"].to(device).long()
        labels["target"] = labels["target"].to(device).long()
        labels["bbox"] = labels["bbox"].to(device)

        optimizer.zero_grad()
        with autocast():
            instrument_logits, verb_logits, target_logits, bbox = model(frames)

            # Compute losses
            loss_instrument = focal_loss_instrument(instrument_logits, labels["instrument"])
            loss_verb = focal_loss_verb(verb_logits, labels["verb"])
            loss_target = focal_loss_target(target_logits, labels["target"])
            loss_bbox = nn.functional.mse_loss(bbox, labels["bbox"])

            # Total loss
            loss = loss_instrument + loss_verb + loss_target + loss_bbox

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
    scheduler.step()

    print(f"Epoch {epoch + 1}, Loss: {loss.item()}")

# Save the trained model
torch.save(model.state_dict(), "best_model.pth")
print("Model saved to best_model.pth")
```

```python
import os
import json
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Custom Dataset Class for Test Videos
class SurgicalTripletTestDataset(Dataset):
    def __init__(self, videos_dir, transform=None):
        self.videos_dir = videos_dir
        self.transform = transform
        self.video_ids = [vid for vid in os.listdir(videos_dir) if os.path.isdir(os.path.join(videos_dir, vid))]
        self.frames = self._load_frames()

    def _load_frames(self):
        frames = []
        for video_id in self.video_ids:
            video_path = os.path.join(self.videos_dir, video_id)
            frame_files = sorted(os.listdir(video_path))
            for frame_file in frame_files:
                frames.append((video_id, frame_file))
        return frames

    def __len__(self):
        return len(self.frames)

    def __getitem__(self, idx):
        video_id, frame_file = self.frames[idx]
        frame_path = os.path.join(self.videos_dir, video_id, frame_file)

        # Load frame
        frame = Image.open(frame_path).convert("RGB")
        if self.transform:
            frame = self.transform(frame)

        # Extract frame ID from filename (e.g., "000000.png" -> "0")
        frame_id = str(int(frame_file.split(".")[0]))  # Convert to integer first, then to string

        return frame, video_id, frame_id

# Define paths
dataset_path = "/content/CholecT50"  # Updated dataset path
test_videos_dir = os.path.join(dataset_path, "videos")  # Test videos directory
labels_dir = os.path.join(dataset_path, "labels")  # Labels directory

# Define transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Create test dataset and dataloader
test_dataset = SurgicalTripletTestDataset(test_videos_dir, transform=transform)
test_dataloader = DataLoader(test_dataset, batch_size=8, shuffle=False)

# Load the trained model
model = SurgicalActionModel(
    num_instruments=0,  # Replace with the actual number of instrument classes
    num_verbs=10,        # Replace with the actual number of verb classes
    num_targets=15,      # Replace with the actual number of target classes
).to(device)

# Load the trained model weights
model_path = "best_model (1).pth"  # Path to the downloaded model
if os.path.exists(model_path):
    print("Loading model from:", model_path)
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.eval()
else:
    print(f"Error: Model file not found at {model_path}. Please train the model first.")
    exit()

# Function to generate predictions
def generate_predictions(model, dataloader):
    predictions = {}
    with torch.no_grad():
        for frames, video_ids, frame_ids in dataloader:
            # Move data to GPU
            frames = frames.to(device)

            # Forward pass
            instrument_logits, verb_logits, target_logits, bbox = model(frames)

            # Convert logits to probabilities
            instrument_probs = torch.softmax(instrument_logits, dim=1)
            verb_probs = torch.softmax(verb_logits, dim=1)
            target_probs = torch.softmax(target_logits, dim=1)

            # Convert bounding box predictions to list format
            bbox = bbox.cpu().numpy().tolist()

            # Format predictions
            for i in range(len(video_ids)):
                video_id = video_ids[i]
                frame_id = frame_ids[i]

                if video_id not in predictions:
                    predictions[video_id] = {}

                predictions[video_id][frame_id] = {
                    "recognition": [
                        instrument_probs[i].cpu().numpy().tolist(),
                        verb_probs[i].cpu().numpy().tolist(),
                        target_probs[i].cpu().numpy().tolist(),
                    ],
                    "detection": [
                        {
                            "triplet": 0,  # Placeholder for triplet ID
                            "instrument": [0, 0, *bbox[i]],  # Placeholder for instrument ID and probability
                        }
                    ],
                }

    return predictions

# Generate predictions for test videos
predictions = generate_predictions(model, test_dataloader)

# Filter predictions for the 5 test videos
test_videos = ["VID92", "VID96", "VID103", "VID110", "VID111"]
filtered_predictions = {video: predictions.get(video, {}) for video in test_videos}

# Save predictions to JSON
output_path = "MODEL_NAME.json"  # Replace MODEL_NAME with your model's name
with open(output_path, "w") as f:
    json.dump(filtered_predictions, f)

print(f"Predictions saved to {output_path}")
```
